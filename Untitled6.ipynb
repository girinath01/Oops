```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
df=pd.read_csv("/content/groceries.csv")
df
```

```python
transaction=df.apply(lambda row:row.dropna().tolist(),axis=1).tolist()
```

```python
te=TransactionEncoder()
te_ary=te.fit(transaction).transform(transaction)
decode=pd.DataFrame(te_ary,columns=te.columns_)
```

```python
frequent=apriori(decode,min_support=0.01,use_colnames=True)
rules=association_rules(frequent,metric="confidence",min_threshold=0.3)
print(rules[["antecedents","consequents","support","confidence","lift"]])
```
